```python
import pandas as pd
import numpy as np
import joblib
```

```python
wk = pd.read_csv('Terragon dataset.csv')
wk.info()
```

```python
wk.sample(15)
```

```python
wk.drop('location_city', axis=1, inplace=True)
```

```python
cat = [i for i in wk.columns if wk[i].dtype == object]
cat
```

```python
num = [p for p in wk.columns if p not in cat]
num
```

```python
wk.isna().sum()
```

```python
# from sklearn.impute import KNNImputer, SimpleImputer
# knim = KNNImputer(n_neighbors=5)
# missing = [i for i in num if wk[i].isna().sum() != 0]
# if missing != []:
#     wk[missing]= knim.fit_transform(wk[missing])

# # wk[missing].fillna('unspecified', inplace=True)

# obj = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
# missing = [i for i in cat if wk[i].isna().sum() != 0]
# if missing != []:
#     wk[missing] = obj.fit_transform(wk[missing])

# # wk[missing].fillna('unspecified', inplace=True)

for a in cat:
    wk[a].fillna(wk[a].mode()[0],inplace=True)
for b in num:
    wk[b].fillna(wk[b].median(), inplace=True)

wk.info()
```

```python
wkd = wk.copy()
wkd.corr()
```

```python
wkd['@timestamp'] = pd.to_datetime(wkd['@timestamp'])
wkd['day'] = [t.date().strftime('%A') for t in wkd['@timestamp']]
wkd['period'] = ['day' if i.time().hour in range(6, 21) else 'night' for i in wkd['@timestamp']]


wkd['age_group'] = ['1 - 12' if i in range(1,13)
                      else '13 - 16' if i in range(13,17)
                      else '17 - 40' if i in range(17,41)
                      else '41 - 64' if i in range(41,65)
                      else '65+'
                      for i in wkd.age]
```

```python
wkd[cat].nunique()
```

```python
unwanted=['msisdn', '@timestamp', 'device_manufacturer','ad_id', 'location_lga', 'ad_name', 'os_name','os_version','location_state'] # 
cat = [i for i in wkd.columns if wkd[i].dtype == object]
nw_cat = list(set(cat) - set(unwanted))
col = list(set(nw_cat).union(set(num)))
col
```

```python
wkd['event_type'].value_counts()
```

```python
c = list(set(nw_cat)- set(['event_type']))
```

```python
wkd = wkd[col]
wkd['event_type'] =  wkd['event_type'].map({'Click':1, 'sms':0})
wkd.head()
```

```python
X = wkd.drop(['event_type', 'spend_vas', 'age'], axis=1)
y = wkd['event_type']

from sklearn.model_selection import train_test_split
trainX, testX, trainy,testy = train_test_split(X, y, test_size=0.25, random_state= 101)
```

```python
trainX.shape, testX.shape
```

```python
for i in c:
    print(f'{i}:{ X[i].unique()}')
```

```python
import category_encoders as ce
binary = ce.BinaryEncoder(cols=c)
trainX = binary.fit_transform(trainX)
joblib.dump(binary, open('binary.pkl', 'wb'))
testX = binary.transform(testX)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
under = RandomUnderSampler()
over = SMOTE()
trainX, trainy = under.fit_resample(trainX, trainy)
trainX, trainy = over.fit_resample(trainX, trainy)
```

```python
trainy.value_counts()
```

```python
from sklearn.preprocessing import PowerTransformer, RobustScaler
sc = RobustScaler()
trainX = sc.fit_transform(trainX)
joblib.dump(sc, open('sc.pkl', 'wb'))
testX = sc.transform(testX)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
rfc = RandomForestClassifier()
scores  = cross_val_score(rfc, trainX, trainy, cv = 10, scoring = 'f1')
print(f"Validation score: {np.mean(scores)}")
rfc.fit(trainX, trainy)
joblib.dump(rfc, open('rfc.pkl', 'wb'))
ypred = rfc.predict(testX)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(ypred, testy))
print(confusion_matrix(ypred, testy))
```

```python
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from imblearn.pipeline import Pipeline


pipe = Pipeline(steps=[('encode', binary),('under', under),('over', over), ('trans', sc), ('model', rfc)])
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipe, X, y, scoring='roc_auc',cv = cv, n_jobs= -1)
print(np.mean(scores), np.std(scores))
```
